```python
import matplotlib.pyplot as plt
import pandas as pd
import folium

df = pd.read_csv('./소상공인시장진흥공단_상가(상권)정보_경기_202203.csv')
```

```python
oliveyoung_suwon = df[(df['도로명주소'].str.contains('수원'))& (df['상호명'].str.contains('올리브영'))]
oliveyoung_hwaseong = df[(df['도로명주소'].str.contains('화성'))& (df['상호명'].str.contains('올리브영'))]

suwon_map = folium.Map(location = [37.23, 127.07], zoom_start = 11.8)


for name, lat, log in zip(oliveyoung_suwon['지점명'], oliveyoung_suwon['위도'], oliveyoung_suwon['경도']):
                          folium.CircleMarker([lat, log],
                                                popup = name,
                                                color = '#ffc2fe',
                                                radius = 15,
                                                fill = True,
                                                fill_opacity = 0.7,
                                                fill_col = '#ffc2dd').add_to(suwon_map)
                          
suwon_map.save('./수원올리브영.html')


```
